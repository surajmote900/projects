# Devide Data into training and testing 

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,Ytrain,Ytest=train_test_split(Xnew,Y,test_size=0.2,random_state=31)
```

# Making Ridge model

```python
from sklearn.linear_model import Ridge
rr=Ridge()
model=rr.fit(Xtrain,Ytrain)
tr_pred=model.predict(Xtrain)
ts_pred=model.predict(Xtest)
from sklearn.metrics import mean_absolute_error
tr_err=mean_absolute_error(Ytrain,tr_pred)
ts_err=mean_absolute_error(Ytest,ts_pred)
print("training_error",tr_err,"\ntest_error",ts_err)
```

```python
X=0.95
for i in range(1,30,1):
    from sklearn.linear_model import Ridge
    rr=Ridge()
    model=rr.fit(Xtrain,Ytrain)
    tr_pred=model.predict(Xtrain)
    ts_pred=model.predict(Xtest)
    from sklearn.metrics import mean_absolute_error
    tr_err=mean_absolute_error(Ytrain,tr_pred)
    ts_err=mean_absolute_error(Ytest,ts_pred)
    print("==============================",X,"==============================")
    print("training_error",tr_err,"\ntest_error",ts_err)
    X=X+0.005
    
```

# making Lasso model

```python
X=0.95
for i in range(0,30,1):
    

    from sklearn.linear_model import Lasso
    ls=Lasso()
    model=ls.fit(Xtrain,Ytrain)
    tr_pred=model.predict(Xtrain)
    ts_pred=model.predict(Xtest)
    from sklearn.metrics import mean_absolute_error
    tr_err=mean_absolute_error(Ytrain,tr_pred)
    ts_err=mean_absolute_error(Ytest,ts_pred)
    X=X+0.0005
    print("===============",X,"=============================")
    print("training_err",tr_err,"test_err",ts_err)
```

# making decision Tree


```python
for i in range(2,30,1):
    from sklearn.tree import DecisionTreeRegressor
    dtr=DecisionTreeRegressor(random_state=31,max_depth=i)
    model=dtr.fit(Xtrain,Ytrain)
    tr_pred=model.predict(Xtrain)
    ts_pred=model.predict(Xtest)
    from sklearn.metrics import mean_absolute_error
    tr_err=mean_absolute_error(Ytrain,tr_pred)
    ts_err=mean_absolute_error(Ytest,ts_pred)
    print("==============================================",i,"================================================")
    print("training_error",tr_err,",testing_error",ts_err)
```
